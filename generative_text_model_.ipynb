```python
# ================================
# 📌 Cell 2: Import Libraries
# ================================
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
# ================================
# 📌 Cell 3: Load Pre-trained Model
# ================================
# Using GPT-2 (lightweight & Colab-friendly)
model_name = "gpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

model.eval()
print("Model loaded successfully!")
```

```python
# ================================
# 📌 Cell 4: Text Generation Function
# ================================
def generate_text(prompt, max_length=150, temperature=0.8):
    inputs = tokenizer(prompt, return_tensors="pt")

    with torch.no_grad():
        output = model.generate(
            inputs["input_ids"],
            max_length=max_length,
            temperature=temperature,
            top_k=50,
            top_p=0.95,
            do_sample=True,
            eos_token_id=tokenizer.eos_token_id
        )

    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
# ================================
# 📌 Cell 5: Example 1 – Simple Prompt
# ================================
prompt = "Artificial Intelligence is changing the future by"
generated_text = generate_text(prompt)

print("Prompt:\n", prompt)
print("\nGenerated Text:\n", generated_text)
```

```python
# ================================
# 📌 Cell 6: Example 2 – Story Generation
# ================================
prompt = "Once upon a time, in a city powered entirely by AI,"
generated_text = generate_text(prompt, max_length=200)

print("Prompt:\n", prompt)
print("\nGenerated Text:\n", generated_text)
```

```python
# ================================
# 📌 Cell 7: Example 3 – Technical Prompt
# ================================
prompt = "Explain machine learning in simple terms:"
generated_text = generate_text(prompt)

print("Prompt:\n", prompt)
print("\nGenerated Text:\n", generated_text)
```

```python
# ================================
# 📌 Cell 8: User Input Prompt
# ================================
user_prompt = input("Enter your prompt: ")
generated_text = generate_text(user_prompt, max_length=180)

print("\nGenerated Text:\n", generated_text)
```
